```python
df = pd.DataFrame(products)
df.columns = ["Nome", "Secao", "Disponivel"]
df
```

```python
# model = "models/embedding-001"
model = "models/embedding-001"

# def embed_fn(title, text, tipo_do_produto):
# content=f"O produto {title} é encontrado na {text} e é um tipo de {tipo_do_produto}.",  # Inclua o tipo do produto
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                   content=f"O produto {title} é encontrado na {text}.",  # Inclua o tipo do produto
                                   title=title,
                                   task_type="RETRIEVAL_DOCUMENT")["embedding"]


# def embed_fn(title, text):
#   return genai.embed_content(model=model,
#                                 content=text,
#                                 title=title,
#                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Nome"], row["Secao"]), axis=1)

df
```

```python
def generate_and_search(consulta, base, model):
  consulta_embedding = genai.embed_content(model=model,
                                content=consulta,
                                task_type="RETRIEVAL_DOCUMENT")["embedding"]

  # Calcula a similaridade usando a distância do cosseno
  similaridades = cosine_similarity([consulta_embedding], np.stack(df["Embeddings"]))

  # index = np.argmax(produto_escalares)
  # Encontra o índice do produto mais similar
  index = np.argsort(similaridades)[0][::-1]
  index = index[0]  # Pega o índice do produto mais similar

  # return df.iloc[index]["Secao"]
  return df.iloc[index]
```

```python
# consulta = "Onde posso encontrar café?"
# consulta = "Onde posso encontrar arroz?"
consulta = str(input("Olá, em que posso ajudar? "))

produto_encontrado = generate_and_search(consulta, df, model)

produto_encontrado
```

```python
generation_config = {
    "temperature": 0.4,
    "candidate_count": 1,
}
```
